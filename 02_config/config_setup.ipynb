## Table Paths

Here I'm setting up the paths to our four main data tables in the workspace catalog. If the table names or locations change, just update them here.

```python
# Databricks Delta table paths
BILLINGS_TABLE      = "workspace.`ds-raw-datasets`.raw_billings"
CC_CALLS_TABLE      = "workspace.`ds-raw-datasets`.raw_cc_calls"
EMAILS_TABLE        = "workspace.`ds-raw-datasets`.raw_emails"
RENEWAL_CALLS_TABLE = "workspace.`ds-raw-datasets`.raw_renewal_calls"
```

## Constants

These are the key parameters for our churn definition and model training. The 28-day threshold is important because it defines when we consider a late renewal to be a churn.

```python
# Churn definition parameter
# A customer who closes more than 28 days after their renewal date is considered churned
CHURN_DAYS_THRESHOLD = 28

# Model training year (using 2023 as it has the highest churn rate)
TRAINING_YEAR = 2023

# Years to analyze for target population
TARGET_YEARS = [2023, 2024, 2025]
```

Just printing out the config to confirm everything loaded correctly.
