## Imports

Bringing in matplotlib and seaborn for plotting.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
%matplotlib inline
```

## Plot Theme Setup

Setting up a custom dark theme for all our charts. The colors are chosen to work well in Databricks notebooks and make the data stand out.

```python
# Apply dark theme styling
plt.rcParams.update({
    'figure.facecolor': '#0a0e1a',
    'axes.facecolor': '#111827',
    'axes.edgecolor': '#1e2d45',
    'axes.labelcolor': '#94a3b8',
    'axes.titlecolor': '#e2e8f0',
    'xtick.color': '#64748b',
    'ytick.color': '#64748b',
    'text.color': '#e2e8f0',
    'grid.color': '#1e2d45',
    'grid.alpha': 0.5,
    'figure.titlesize': 15,
    'axes.titlesize': 12,
    'axes.labelsize': 10,
    'legend.facecolor': '#1a2236',
    'legend.edgecolor': '#1e2d45',
})
```

## Color Palette

Our standard color palette for different chart elements. Using semantic names so it's clear what each color represents.

```python
# Color palette for consistent styling
PALETTE = {
    'danger': '#ef4444',   # Red for high churn/alerts
    'amber': '#f59e0b',    # Orange for warnings/medium
    'blue': '#3b82f6',     # Blue for neutral/info
    'green': '#10b981',    # Green for good/low churn
    'muted': '#334155'     # Gray for background elements
}
```

## Color Helper Functions

These functions return the right color based on year or churn rate. Makes it easier to keep charts consistent without having to remember which color goes where.

```python
def year_color(year):
    """
    Get color for a specific year.
    2023 = red (peak churn), 2024 = amber, 2025 = green (improving)
    
    Args:
        year: Year as integer
    
    Returns:
        Hex color code
    """
    year_colors = {
        2023: PALETTE['danger'],
        2024: PALETTE['amber'],
        2025: PALETTE['green']
    }
    return year_colors.get(year, PALETTE['blue'])

def rate_color(rate):
    """
    Get color based on churn rate threshold.
    High (>=12%) = red, Medium (>=9%) = amber, Low (<9%) = green
    
    Args:
        rate: Churn rate as percentage
    
    Returns:
        Hex color code
    """
    if rate >= 12:
        return PALETTE['danger']
    if rate >= 9:
        return PALETTE['amber']
    return PALETTE['green']
```
